```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urlparse

def extract_school_name_from_homepage(url):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, 'html.parser')
            # Modify the following code to match the HTML structure of the school name on the homepage
            school_name_element = soup.find('h1', class_='school-name')
            if school_name_element:
                school_name = school_name_element.text.strip()
                return school_name
            else:
                print(f"School name element not found on the homepage of {url}")
        else:
            print(f"Unable to access URL: {url}")
    except requests.exceptions.RequestException as e:
        print(f"Error accessing URL: {url}")
        print(e)
    return None

# Example homepage URLs
homepage_urls = [
    "http://312.clayton.k12.ga.us/",
    "http://EDON-NW.ORG",
    "http://EKACADEMY.ORG"
]

for url in homepage_urls:
    school_name = extract_school_name_from_homepage(url)
    if school_name:
        print(f"School name for {url}: {school_name}")
```

```python
import requests
from bs4 import BeautifulSoup
from transformers import T5ForConditionalGeneration, T5Tokenizer

# Load the model and tokenizer
model_name = "google/flan-t5-small"
model = T5ForConditionalGeneration.from_pretrained(model_name)
tokenizer = T5Tokenizer.from_pretrained(model_name)

# Define the URL
url = "https://www.smore.com/84mkp"

# Send a GET request to the URL and extract the HTML content
response = requests.get(url)
html_content = response.content

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

# Extract the name and email based on the HTML structure of the webpage
name_element = soup.find('span', class_='header-name')
email_element = soup.find('a', class_='header-email')

# Process the extracted text
name = name_element.text.strip() if name_element else None
email = email_element.text.strip() if email_element else None

# Print the extracted name and email
print("Name:", name)
print("Email:", email)
```
