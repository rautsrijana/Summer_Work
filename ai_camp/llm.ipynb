```python
from bs4 import BeautifulSoup
import requests
url = "https://animashighschool.com/faculty-staff/"
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.content, "html.parser")
for style in soup.find_all('style'):
    style.decompose()

for script in soup.find_all('script'):
    script.decompose()

modified_html = str(soup)
output_file = 'html-trim.html'
with open(output_file, 'w', encoding='utf-8') as file:
    file.write(modified_html)
```

```python
from langchain.document_loaders import BSHTMLLoader
from langchain.chat_models import ChatOpenAI
from langchain.chains.qa_with_sources import load_qa_with_sources_chain


loader = BSHTMLLoader("html-trim.html")
data = loader.load()

llm = ChatOpenAI(model = 'gpt-3.5-turbo', temperature=0, openai_api_key=API_KEY)
# prompt = "You are developing a program to extract staff information from school directory pages. Given a list of URLs for different school websites, you want to extract details such as names, positions, and contact information for the staff members listed on their staff directory pages. Your goal is to create a script that can efficiently scrape this information from each URL and store it in a structured format. Use web scraping techniques to extract staff information from the provided URLs. Consider handling different HTML structures and variations across school websites. Feel free to include any necessary libraries or frameworks to accomplish this task effectively."
# prompt = "You are developing a program to extract staff information from school directory pages. Given a list of URLs for different school websites, you want to extract details such as names, positions, and contact information for the staff members listed on their staff directory pages. Your goal is to create a script that can efficiently scrape this information from each URL and store it in a structured format."
prompt = "You need to extract staff information from school directory pages. Given a list of URLs for different school websites"

chain = load_qa_with_sources_chain(llm, chain_type = "stuff")
print("Extracting Data......")

output = chain.run(input_documents = data, question = prompt)
print(output)
```

```python
import csv
# Process the extracted data
staff_data = output.split(";")
staff_records = [record.strip().split(", ") for record in staff_data]

# Save the data into a CSV file
filename = "staff_data.csv"
with open(filename, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Name', 'Email'])
    writer.writerows(staff_records)

print("Data saved into", filename)
```
