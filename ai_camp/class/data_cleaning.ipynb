```python
def separate_paragraphs(file_path):
    # Read the contents of the text file
    with open(file_path, 'r') as file:
        content = file.read()

    # Split the content into paragraphs based on double line breaks
    paragraphs = content.split('\n\n\n\n\n\n\n')

    # Join the paragraphs with a large blank line between them
    separated_content = '\n\n\n\n\n\n\n\n\n\n'.join(paragraphs)

    # Write the separated content to a new file
    new_file_path = 'separated_file.txt'
    with open(new_file_path, 'w') as file:
        file.write(separated_content)

    print("Paragraphs separated and saved to '{}'.".format(new_file_path))

# Specify the path to your input text file
input_file_path = 'combined_data.txt'

# Call the function to separate paragraphs
separate_paragraphs(input_file_path)
```
